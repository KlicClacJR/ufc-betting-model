I got the fight data from: https://raw.githubusercontent.com/komaksym/UFC-DataLab/refs/heads/main/data/stats/stats_raw.csv
It contains the stats from every fight in the UFC up to 3/14/2026, such as: fighter names, date, nicknames, result of the fight (winner, method, round and time), referee, division, event name and location, and fighting stats(strikes, takedowns, etc.)

I got the fighter statistics from: https://github.com/komaksym/UFC-DataLab/blob/main/data/external_data/raw_fighter_details.csv 
It contains: Name, height, weight, reach, stance. and fighting statistics (takedown defense, striked per minute, etc.)


The first issue I ran into was in the information barrier. The database of the fighter statistics had the career statistics for each fighter. For example, if we consider a fight in 2022, the dataset today includes statistics from the fighter’s entire career, including fights after 2022. However, at the time of that fight, only data from before 2022 would have been available. This creates a problem, because using the full dataset would introduce information that would not have been known at the time, leading to unrealistic or biased predictions. This introduces a form of data leakage, because using the full dataset is used to predict past events.
How can I ensure when building the model we only have access to the data from before that date? 
I decided to separate the fighter statistics based on when the information became available. For instance, for each fight, we will combine the statistics known before the fight into that fight, to ensure the data leakage is not a problem.
To do this, I decided to sort the fights by date, and all the information available before the fight would be figured out by adding their data from the previous fights, and age can be calculated using their birth day.
Assumptions: Fighter height/weight/reach did not change (will address weight division changes later), and fighter stance was consistent throughout their careers. This may not necessarily be true, as many fighters change stances/switch between stances throughout their careers. 
Therefore, the first step is to clean and structure the data before building any models.
